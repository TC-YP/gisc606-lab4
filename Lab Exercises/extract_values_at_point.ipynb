```python
import os
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import numpy.ma as ma
import pandas as pd
import rioxarray as rxr
from rasterio.plot import plotting_extent
import geopandas as gpd

import rasterstats as rs
import earthpy as et
import earthpy.plot as ep
```

```python
sns.set_style("white")
sns.set(font_scale=1.5)
```

```python
data = et.data.get_data("spatial-vector-lidar")
os.chdir(os.path.join(et.io.HOME, 
                      'earth-analytics', 
                      'data'))
```

```python
sjer_lidar_chm_path = os.path.join("spatial-vector-lidar",
                                   "california", 
                                   "neon-sjer-site",
                                   "2013", 
                                   "lidar", 
                                   "SJER_lidarCHM.tif")

sjer_chm_data = rxr.open_rasterio(sjer_lidar_chm_path, masked=True).squeeze()
```

```python
ax = ep.hist(sjer_chm_data.values,
             figsize=(8, 8),
             colors="teal",
             xlabel="Lidar Estimated Tree Height (m)",
             ylabel="Total Pixels",
             title="Distributions of Pixel Values \n Lidar Canopy Height Model")

ax[1].ticklabel_format(useOffset=False,
                         style='plain')
```

```python
schmd = sjer_chm_data
```

```python
print('Mean:', schmd.mean().values)
print('Max:', schmd.max().values)
print('Min:', schmd.min().values)
```

```python
schmd_no_zeros = schmd.where(schmd != 0, np.nan)

print('Mean:', schmd_no_zeros.mean().values)
print('Max:', schmd_no_zeros.max().values)
print('Min:', schmd_no_zeros.min().values)
```

```python
ax = ep.hist(schmd_no_zeros.values,
             figsize=(8, 8),
             colors="teal",
             xlabel="Lidar Estimated Tree Height (m)",
             ylabel="Total Pixels",
             title="Distributions of Pixel Values \n Lidar Canopy Height Model")

ax[1].ticklabel_format(useOffset=False,
                         style='plain')
```

```python
sjer_centroids_path = os.path.join("spatial-vector-lidar",
                                   "california", 
                                   "neon-sjer-site",
                                   "vector_data", 
                                   "SJER_plot_centroids.shp")

sjer_plts_pnts = gpd.read_file(sjer_centroids_path)

type(sjer_plts_pnts)
```

```python
sjer_plts_pnts.geom_type.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

ep.plot_bands(schmd,
             extent=plotting_extent(schmd,
                                    schmd.rio.transform()),
             cmap='Blues',
             title="San Joachin Field Site \n Vegetation Plot Locations",
              scale=False,
              ax=ax)

sjer_plts_pnts.plot(ax=ax,
                       marker='s',
                       markersize=45,
                       color='purple')
ax.set_axis_off()
plt.show()
```

```python
sjer_plts_poly = sjer_plts_pnts.copy()

sjer_plts_poly["geometry"] = sjer_plts_pnts.geometry.buffer(20)
sjer_plts_poly.head()
```

```python
output_path = os.path.join("spatial-vector-lidar", 
                           "outputs")

if not os.path.isdir(output_path):
    os.mkdir(output_path)

plot_buffer_path = os.path.join(output_path, 
                                "plot_buffer.shp")

sjer_plts_poly.to_file(plot_buffer_path)
```

```python
sjer_tree_hghts = rs.zonal_stats(plot_buffer_path,
                                   schmd_no_zeros.values,
                                   nodata=-999,
                                   affine=schmd_no_zeros.rio.transform(),
                                   geojson_out=True,
                                   copy_properties=True,
                                   stats="count min mean max median")

type(sjer_tree_hghts)
```

```python
sjer_lidar_hght_df = gpd.GeoDataFrame.from_features(sjer_tree_hghts)
sjer_lidar_hght_df.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

ax.bar(sjer_lidar_hght_df['Plot_ID'],
       sjer_lidar_hght_df['max'],
       color="teal")

ax.set(xlabel='Plot ID', 
       ylabel='Max Height',
       title='Maximum LIDAR Derived Tree Heights')

plt.setp(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.show()
```

```python
sjer_tree_hghts_ras = rs.zonal_stats(plot_buffer_path,
                                     schmd_no_zeros.values,
                                     nodata=-999,
                                     affine=schmd.no_zeros.rio.transform(),
                                     geojson_out=True,
                                     raster_out=True,
                                     copy_properties=True,
                                     stat="count min mean max median")

sjer_lidar_hght_df_ras = gpd.GeoDataFrame.from_features(sjer_tree_hghts_ras)

sjer_lidar_hght_df_ras[["Plot_ID", 
                          "count", 
                          "geometry",
                          "mini_raster_affine", 
                          "mini_raster_array"]].head()
```
